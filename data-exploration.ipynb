```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Let us plot missing values to check for patterns/bias
# ax = plt.axes() #no need to do this is you're doing plt.figure
plt.figure(figsize=(15, 5))
sns.heatmap(books.isna().transpose(), cbar=False)
plt.xlabel("Rows")
plt.ylabel("Columns with missing values")
```

3. PATTERN -> Looks like average_rating, num_pages and ratings_count are all null when one of them is null (perhaps obtained from separate dataset?? where all books in bigger dataset were not contained)
4. Look closer into description missing values

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_the_book"] = 2025 - books["published_year"]
```

```python
#Let us check if any of these columns cause missing descriptions
columns_of_interest = ["num_pages", "age_of_the_book", "missing_description", "average_rating"]
```

```python
correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
heatmap.set_title("Correlation HeatMap")
heatmap
```

```python
cleaned_books = books[~(books["description"].isna()) &
                      ~(books["num_pages"].isna()) &
                      ~(books["average_rating"].isna()) &
                      ~(books["published_year"].isna())
                      ]
cleaned_books
```

```python
cleaned_books_25_words = cleaned_books[cleaned_books["words_in_description"] >= 25]
cleaned_books_25_words
```

```python
cleaned_books_25_words["title_and_subtitle"] = (
    np.where(cleaned_books_25_words["subtitle"].isna(),
             cleaned_books_25_words["title"],
             cleaned_books_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
cleaned_books_25_words
```

```python
#To make an unique identifier for description
cleaned_books_25_words["tagged_desc"] = cleaned_books_25_words["isbn13"].astype(str).str.strip() + " " + \
                                        cleaned_books_25_words["description"].astype(str).str.strip()
```

```python
cleaned_books_25_words
```

```python
(cleaned_books_25_words
 .drop(["subtitle", "missing_description", "age_of_the_book", "words_in_description"], axis=1)
 .to_csv("books_cleaned.csv", index=False)
 )
```
